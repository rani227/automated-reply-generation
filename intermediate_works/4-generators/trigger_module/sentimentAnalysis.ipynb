```python
# Load your file
df = pd.read_csv("trigger_output.csv")   # change filename if needed
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
# Apply sentiment analysis
def get_sentiment_score(text):
    if pd.isna(text):
        return 0   # neutral for missing text
    score = analyzer.polarity_scores(text)["compound"]
    return score

df["review_sentiment"] = df["cleaned_review2"].apply(get_sentiment_score)
```

```python
# Convert numeric score to label
def sentiment_label(score):
    if score > 0:
        return "positive"
    elif score < 0:
        return "negative"
    else:
        return "neutral"

df["sentiment_label"] = df["review_sentiment"].apply(sentiment_label)
```

```python
# Save updated file
df.to_csv("/content/sentiment_output.csv", index=False)

df.head()
```
