```python
import pandas as pd
df = pd.read_csv("routing_output.csv")
```

```python
df.columns
```

```python
pos_short_df = df[df["final_generator"] == "pos_short"]
pos_long_df  = df[df["final_generator"] == "pos_long"]
neg_short_df = df[df["final_generator"] == "neg_short"]
neg_long_df  = df[df["final_generator"] == "neg_long"]
```

```python
pos_short_df.to_csv("/content/pos_short.csv", index=False)
pos_long_df.to_csv("/content/pos_long.csv", index=False)
neg_short_df.to_csv("/content/neg_short.csv", index=False)
neg_long_df.to_csv("/content/neg_long.csv", index=False)
```

```python
print("pos_short:", len(pos_short_df))
print("pos_long:", len(pos_long_df))
print("neg_short:", len(neg_short_df))
print("neg_long:", len(neg_long_df))
```
